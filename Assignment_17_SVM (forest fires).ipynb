```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
# find categorical variables in training data set

traincategorical = [var for var in df.columns if df[var].dtype=='O']

print('There are {} categorical variables\n'.format(len(traincategorical)))

print('The categorical variables are :\n\n', traincategorical)
```

```python
X= df.iloc[:,:-1]
y=df.iloc[:,-1]

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X=X.apply(LabelEncoder().fit_transform)
#sns.boxplot(x="size_category",y="DC",data=fire)
```

```python
X_ = X.iloc[:,0:29]
```

```python
X_
```

```python
y
```

```python
# Normalizing the data
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)

X_norm = norm_func(X_)
X_norm
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_norm,y,test_size = 0.25, stratify = y)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(X_train,y_train)
pred_test = model_linear.predict(X_test)
np.mean(pred_test==y_test)
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(X_train,y_train)
pred_test_rbf = model_rbf.predict(X_test)
np.mean(pred_test_rbf==y_test)
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(X_train,y_train)
pred_test_poly = model_poly.predict(X_test)

np.mean(pred_test_poly==y_test)
```

```python
#'sigmoid'
model_sig = SVC(kernel = "sigmoid")
model_sig.fit(X_train,y_train)
pred_test_sig = model_sig.predict(X_test)

np.mean(pred_test_sig==y_test)
```

```python
model=SVC()
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
print(confusion_matrix(pred,y_test))
print(classification_report(pred,y_test))
```

```python
np.mean(pred==y_test)*100
```

```python
acc = accuracy_score(y_test, pred) * 100
print("Accuracy =", acc)
```
